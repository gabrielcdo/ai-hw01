# Data Visualization 

```python
import os
import cv2
import matplotlib.pyplot as plt
import random


# Function to display one random example image from a directory
def display_random_example(directory, category, ax):
    image_files = [f for f in os.listdir(directory) if f.endswith(".jpg")]

    if image_files:
        # Select a random image from the directory
        random_image = random.choice(image_files)
        img_path = os.path.join(directory, random_image)
        img = cv2.imread(img_path)
        img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

        ax.imshow(img_rgb)
        ax.set_title(f"{category}")
        ax.axis("off")
    else:
        print(f"No images found in {directory}")


# Specify the directories
adidas_dir = "../data/raw_image/adidas_dir"
adidas_esq = "../data/raw_image/adidas_esq"
fila_dir = "../data/raw_image/fila_dir"
fila_esq = "../data/raw_image/fila_esq"
social_dir = "../data/raw_image/social_dir"
social_esq = "../data/raw_image/social_esq"

# Create a 2x3 grid
fig, axs = plt.subplots(2, 3, figsize=(12, 8))
fig.suptitle("Random Example Images", fontsize=16)

# Display one random example image from each category in the grid
display_random_example(adidas_dir, "adidas_dir", axs[0, 0])
display_random_example(adidas_esq, "adidas_esq", axs[0, 1])
display_random_example(fila_dir, "fila_dir", axs[0, 2])
display_random_example(fila_esq, "fila_esq", axs[1, 0])
display_random_example(social_dir, "social_dir", axs[1, 1])
display_random_example(social_esq, "social_esq", axs[1, 2])

plt.show()
```

```python
# do a bar plot of the number of images in each category
plt.figure(figsize=(8, 6))
plt.bar(
    ["adidas_dir", "adidas_esq", "fila_dir", "fila_esq", "social_dir", "social_esq"],
    [
        len(os.listdir(adidas_dir)),
        len(os.listdir(adidas_esq)),
        len(os.listdir(fila_dir)),
        len(os.listdir(fila_esq)),
        len(os.listdir(social_dir)),
        len(os.listdir(social_esq)),
    ],
)
# Set the title and axis label
plt.title("Number of Images in Each Category")
plt.xlabel("Category")
plt.ylabel("Number of Images")
```
